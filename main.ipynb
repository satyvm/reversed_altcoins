```python
def t_p(t_change):
    return str(int(l_ot_BTCUSDT) + t_change - 1)
```

```python
df_h_coins = dict()
all_t_p = {t_p(60000): Client.KLINE_INTERVAL_1MINUTE, t_p(180000): Client.KLINE_INTERVAL_3MINUTE}
all_t_p
```

```python
tp = t_p(86400001)
tp
```

```python
for index, symbol in df_symbol_coins.items():
    if client.get_symbol_info(symbol) is not None:
        h_i_coins = pd.DataFrame.from_dict(client.get_historical_klines(symbol, Client.KLINE_INTERVAL_1MINUTE, l_ot_BTCUSDT, tp))
        if h_i_coins.empty:
            df_h_coins[symbol] = None
            print(symbol, " does not gave a proper output")
            continue
        else:
            h_i_coins.columns = ['OpenTime', 'Open', 'High', 'Low', 'Close', 'Volume', 'CloseTime', 'QuoteAssetVolume', 'NumberOfTrades', 'TB_BaseVolume', 'TB_QuoteVolume', 'Ignore']
            df_h_coins[symbol] = h_i_coins
            print(symbol, " has been added")
    else:
        df_h_coins[symbol] = None
        print(symbol, " does not exist")
```

```python
df_h_coins["ETHUSDT"]
```

```python
def p_change(coin_symbol, time_period):
    close_price = float(df_h_coins[coin_symbol]['Close'].iloc[time_period - 1])
    open_price = float(df_h_coins[coin_symbol]['Open'].iloc[0])
    return ((close_price - open_price)/open_price)*100
p_change("ETHUSDT", 1)
```

```python
def max_p_change(time_period):
    i_max_p_change = p_change("BTCUSDT", time_period)
    i_max_symbol = "BTCUSDT"
    for coin_symbol in df_h_coins:
        if df_h_coins[coin_symbol] is not None:
            i_p_change = p_change(coin_symbol, time_period)
            if i_p_change > i_max_p_change:
                i_max_p_change = i_p_change
                i_max_symbol = coin_symbol
    return i_max_p_change, i_max_symbol
```

```python
max_p_change(1)
```

```python
import csv
```

```python
tp_temp = 5
dict_temp = {}
for coin_symbol in df_h_coins:
    if df_h_coins[coin_symbol] is not None:
            i_p_change = p_change(coin_symbol, tp_temp)
            dict_temp[coin_symbol] = i_p_change
```

```python
pd.DataFrame(dict_temp, index=[0]).T.sort_values(by=1, ascending=False).to_csv("temp.csv")
```

```python
with open('temp.csv', 'w') as csvfile:
    writer = csv.DictWriter(csvfile)
    writer.writeheader()
    writer.writerows(dict_temp)
```
